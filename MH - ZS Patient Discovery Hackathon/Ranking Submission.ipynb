```python
import pandas as pd
import numpy as np
```

```python
ss15=pd.read_excel('MH-ZS_fork_ss15_lgb.xlsx') #85.672
sfork10=pd.read_excel('MH-ZS_fork_s10lgb.xlsx') #85.028
from scipy.stats import rankdata #86.059 ranked
```

```python
predict_list=[]
predict_list.append(ss15['outcome_flag'].values)
predict_list.append(sfork10['outcome_flag'].values)


print("Rank averaging on ", len(predict_list), " files")
predictions = np.zeros_like(predict_list[0])
for predict in predict_list:
        predictions = np.add(predictions, rankdata(predict)/predictions.shape[0])  
predictions /= len(predict_list)
```

```python
predictions
```

```python
s=pd.DataFrame({'patient_id':ss15['patient_id'],'outcome_flag':predictions})
s
```

```python
s.to_excel('final_submission.xlsx',index=False) #rank of 10 files
```
